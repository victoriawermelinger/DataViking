```python
#print(Base_Dados['Data de Adesão'].unique())
#print(Base_Dados['Data de Adesão'].dtype)
# Exemplo de conversão manual
Base_Dados['Data de Adesão'] = pd.to_datetime(Base_Dados['Data de Adesão'], errors='coerce')
```

```python
# Conversão para Data
#Base_Dados['Data de Adesão'] = pd.to_datetime(Base_Dados['Data de Adesão'])
#Base_Dados['Data de Adesão'] = pd.to_datetime(Base_Dados['Data de Adesão'], format='%d/%m/%Y')
Base_Dados['Data de Adesão'] = pd.to_datetime(Base_Dados['Data de Adesão'], format='%d/%m/%Y', errors='coerce')


Base_Dados['Data de Adesão'].head()
```

```python
# Extrair o Ano E Mes
Base_Dados['Mes'] = pd.DatetimeIndex(Base_Dados['Data de Adesão']).month 
Base_Dados['Ano'] = pd.DatetimeIndex(Base_Dados['Data de Adesão']).year

Base_Dados.head()
```

```python
# Tabela Analitica 
Analise_Agrupada = Base_Dados.groupby(by=['Pais', 'Ano','Mes','Empresa']).count()['Setor'].reset_index()
Analise_Agrupada
```

```python
# Localizando o Brazil 
Analise_Agrupada.loc[Analise_Agrupada['Pais']=='Brazil']
```

```python
# Localizando o Estados Unidos 
Analise_Agrupada.loc[Analise_Agrupada['Pais']=='United States']
```

```python
# Transformando a coluna Valor
# Base_Dados['Valor ($)'].apply( lambda Linha: Linha.replace('$', ''))
# Base_Dados['Valor ($)'] = Base_Dados['Valor ($)'].astype(str).str.replace('$', '')
Base_Dados['Valor ($)'] = pd.to_numeric( Base_Dados['Valor ($)'].apply( lambda Linha: Linha.replace('$', '') ) )

Base_Dados.head()
```

```python
# Tabela Analitica 
#Base_Dados.groupby(by=['Pais', 'Setor']).sum()['Valor ($)'].reset_index()
Analise_Pais = Base_Dados.groupby(by=['Pais', 'Setor','Valor ($)']).count()['Ano'].reset_index()

Analise_Pais.head()
```

```python
Analise_Valor = Analise_Pais.sort_values('Valor ($)', ascending=False)
Analise_Valor.head()
```

```python
plt.figure( figsize=(15,6) )
plt.plot( Analise_Valor['Pais'], Analise_Valor['Valor ($)'] )
plt.title('Analise do Valor por Pais')
plt.xticks( rotation=50, ha='right');
```
