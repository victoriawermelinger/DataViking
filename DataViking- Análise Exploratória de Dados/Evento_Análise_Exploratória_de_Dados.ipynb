Projeto: Calculadora de Juros

```python
def calcular_juros_simples( principal, taxa, tempo ):
  juros = principal * ( taxa / 100 ) * tempo
  return juros
```

```python
# Coletar os dados
principal = float( input('Digite o valor principal (R$): ') )
taxa = float(input('Digite a taxa de juros (%): '))
tempo = float(input('Digite o tempo (em anos): '))

# Calcular e exibir os juros
juros_simples = calcular_juros_simples(principal, taxa, tempo)

print(f'Juros Simples: R${juros_simples:.2f}')
```

## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Dados para o gráfico
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
```

```python
# Criar o gráfico de linhas
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Seno de x')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de Linhas')
plt.legend()
```

```python
# Dados para o gráfico
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [5, 7, 8, 4, 6]

# Criar o gráfico de barras
plt.figure(figsize=(8, 6))

plt.bar(categorias, valores, color='skyblue')

plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.title('Gráfico de Barras')
```

```python
# Dados para o gráfico
x = np.random.rand(50)
y = np.random.rand(50)

# Criar o gráfico de dispersão
plt.figure(figsize=(8, 6))

plt.scatter(x, y, color='purple')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de Dispersão')
```

```python
# Dados para o gráfico
dados = np.random.randn(1000)

# Criar o histograma
plt.figure(figsize=(8, 6))

plt.hist(dados, bins=30, color='green', edgecolor='black')

plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title('Histograma')
```

```python
# Dados para o gráfico
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # Explode o primeiro pedaço

# Criar o gráfico de pizza
plt.figure(figsize=(8, 6))

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

plt.title('Gráfico de Pizza')
```

## Seaborn

```python
import seaborn as sns
sns.set(style="darkgrid")
```

```python
plt.figure(figsize=(8, 6))

sns.lineplot(x=x, y=y, label='Seno de x')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de Linhas com Seaborn')
plt.legend()
```

```python
# Criar o gráfico de barras com Seaborn
plt.figure(figsize=(8, 6))

sns.barplot(x=categorias, y=valores, palette='muted')

plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.title('Gráfico de Barras com Seaborn')
```

```python
# Criar o gráfico de dispersão com Seaborn
plt.figure(figsize=(8, 6))

sns.scatterplot(x=x, y=y, color='red')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de Dispersão com Seaborn')
```

```python
# Criar o histograma com Seaborn
plt.figure(figsize=(8, 6))

sns.histplot(dados, bins=30, color='blue', kde=True)

plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title('Histograma com Seaborn')
```

```python
# Dados para o gráfico
np.random.seed(10)
dados = np.random.normal(size=(10, 4)) + np.arange(4) * 2

# Criar o boxplot com Seaborn
plt.figure(figsize=(8, 6))

sns.boxplot(data=dados, palette='Set3')

plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.title('Boxplot com Seaborn')
```
